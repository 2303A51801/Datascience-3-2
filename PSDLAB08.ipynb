```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
data = pd.read_csv('monthly_rainfall.csv')

# Define seasons for each month (India-based example)
season_colors = {
    'Winter': 'skyblue',
    'Summer': 'orange',
    'Monsoon': 'green',
    'Post-Monsoon': 'purple'
}

# Map months to seasons
season_map = {
    'January': 'Winter', 'February': 'Winter',
    'March': 'Summer', 'April': 'Summer', 'May': 'Summer',
    'June': 'Monsoon', 'July': 'Monsoon', 'August': 'Monsoon', 'September': 'Monsoon',
    'October': 'Post-Monsoon', 'November': 'Post-Monsoon', 'December': 'Winter'
}

# Assign colors based on season
data['Season'] = data['Month'].map(season_map)
data['Color'] = data['Season'].map(season_colors)

# Plot bar chart
plt.figure(figsize=(10, 6))
plt.bar(data['Month'], data['Rainfall'], color=data['Color'])

# Chart details
plt.title('Average Monthly Rainfall', fontsize=16, fontweight='bold')
plt.xlabel('Month', fontsize=12)
plt.ylabel('Rainfall (mm)', fontsize=12)
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Add legend manually
handles = [plt.Rectangle((0,0),1,1, color=color) for color in season_colors.values()]
labels = season_colors.keys()
plt.legend(handles, labels, title='Season')

# Annotate rainfall values on top of bars
for i, val in enumerate(data['Rainfall']):
    plt.text(i, val + 5, f"{val}", ha='center', fontsize=9)

plt.tight_layout()
plt.show()

# Insights
wettest = data.loc[data['Rainfall'].idxmax()]
driest = data.loc[data['Rainfall'].idxmin()]

print("üåßÔ∏è Wettest Month:", wettest['Month'], "-", wettest['Rainfall'], "mm")
print("‚òÄÔ∏è Driest Month:", driest['Month'], "-", driest['Rainfall'], "mm")
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load dataset
data = pd.read_csv("traffic_source.csv")

# Define custom colors for each source
source_colors = {
    'Direct': '#1f77b4',     # Blue
    'Search': '#2ca02c',     # Green
    'Social': '#d62728',     # Red
    'Referral': '#9467bd',   # Purple
    'Ads': '#ff7f0e'         # Orange
}

# Assign color to each bar
data['Color'] = data['Source'].map(source_colors)

# Plot bar chart
plt.figure(figsize=(8, 5))
bars = plt.bar(data['Source'], data['Visits'], color=data['Color'])

# Chart details
plt.title("Website Traffic by Source", fontsize=16, fontweight='bold')
plt.xlabel("Traffic Source", fontsize=12)
plt.ylabel("Number of Visits", fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.6)

# Annotate visit numbers on bars
for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, height + 100, f"{height}",
             ha='center', va='bottom', fontsize=10)

plt.tight_layout()
plt.show()

# Insights
top_source = data.loc[data['Visits'].idxmax()]
least_source = data.loc[data['Visits'].idxmin()]

print("üìà Top Traffic Source:", top_source['Source'], "-", top_source['Visits'], "visits")
print("üìâ Least Traffic Source:", least_source['Source'], "-", least_source['Visits'], "visits")
```
